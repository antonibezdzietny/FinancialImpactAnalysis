## Scrap reports
---
The data from the reports are taken from https://www.biznesradar.pl/

```python
import pandas as pd
import numpy as np

wigs_list = pd.read_csv("../database/gpwCompaniesLists/WIGs.csv")
wigs_list.fillna("") # Only for display
```

### Annual Reports

```python
import scripts as scr
import os.path
import time

fr = scr.FinancialReportScraper()
save_path = "../database/annualReports/"

# Prepare list of ticker to scrape
scraper_ticker_list = wigs_list.to_numpy().flatten()
scraper_ticker_list = scraper_ticker_list[~pd.isnull(scraper_ticker_list)]

for ticker in scraper_ticker_list:
    # Continue if data exist
    if os.path.isfile(f"{save_path}{ticker}_Y_D.csv") and \
    os.path.isfile(f"{save_path}{ticker}_Y_V.csv"):
        continue

    time.sleep(1)

    fr.parse(ticker, 
            scr.FinancialReportScraper.ReportType.R, 
            save_path=save_path)
```
